```r
tapply(statedata$HS.Grad, statedata$state.region, mean)
```

```r
boxplot(Murder ~ state.region, statedata)
```

```r
t = subset(statedata, state.region=='Northeast')
tail(t[with(t, order(Murder)), ])
```

```r
model = lm(Life.Exp ~ Population + Income + Illiteracy + Murder + HS.Grad + Frost + Area,
           data=statedata)
summary(model)
```

```r
plot(statedata$Income, statedata$Life.Exp)
```

```r
model = lm(Life.Exp ~ Population + Murder + HS.Grad + Frost,
           data=statedata)
summary(model)
```

```r
head(sort(predict(model)))
statedata[which.min(statedata$Life.Exp), ]
```
