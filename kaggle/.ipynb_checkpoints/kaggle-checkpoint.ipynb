```r
train = read.csv('data/train2016.csv.xz')
test = read.csv('data/test2016.csv.xz')
```

```r
SimpleMod = glm(Party ~ . -USER_ID, data=train,
                family=binomial)
```

```r
PredTest = predict(SimpleMod, newdata=test,
                   type='response')
threshold = 0.5
PredTestLabels = as.factor(ifelse(PredTest < threshold,
                                  'Democrat', 'Republican'))
```

```r
#PredTestLabels
MySubmission = data.frame(USER_ID=test$USER_ID, Predictions=PredTestLabels)
write.csv(MySubmission, 'data/SummissionSimgleLog.csv', row.names=F)
```

```r
Forest = randomForest(Party ~ YOB + Gender + Income + HouseholdStatus +
                      EducationLevel - USER_ID +
                      Q109367 + Q106388 + Q108754 + Q109244 +
                      Q108950 + Q109367 + Q112512 + Q113181 +
                      Q113583 + Q113584 + Q113992 + Q115611 +
                      Q116441 + Q118117 + Q123464 + Q123621, data=train_imputed,
                      ntree=200, nodesize=20)
PredForest = predict(Forest, newdata=test_imputed)
#head(PredForest)
t = table(test_imputed$Party, PredForest)
t
sum(diag(t))/sum(t)
```

```r
test_with_party = test
test_with_party$Party = NA
#dim(test)
#dim(test_with_party)
#dim(train)
combined = rbind(train, test_with_party)
dim(combined)
dim(test)
dim(train)
```

```r
imputed_combined = complete(mice(combined))
```

```r
#uncombined_train = 
uncombined_train = imputed_combined[imputed_combined$USER_ID == train$USER_ID,]
uncombined_test = imputed_combined[imputed_combined$USER_ID == test$USER_ID,]
dim(uncombined_test)
```

```r
spl = sample.split(uncombined_train$Party == 'Democrat', SplitRatio = 0.7)
uncombined_train_train = subset(uncombined_train, spl == T)
uncombined_train_validation = subset(uncombined_train, spl == F)
```

```r
Forest = randomForest(Party ~ YOB + Gender + Income + HouseholdStatus +
                      EducationLevel - USER_ID +
                      Q109367 + Q106388 + Q108754 + Q109244 +
                      Q108950 + Q109367 + Q112512 + Q113181 +
                      Q113583 + Q113584 + Q113992 + Q115611 +
                      Q116441 + Q118117 + Q123464 + Q123621, data=uncombined_train_train,
                      ntree=200, nodesize=20)
PredForest = predict(Forest, newdata=uncombined_train_validation)
#head(PredForest)
t = table(uncombined_train_validation$Party, PredForest)
t
sum(diag(t))/sum(t)
```

```r
PredTest = predict(Forest, newdata=uncombined_test)
MySubmission = data.frame(USER_ID=uncombined_test$USER_ID, Predictions=PredTest)
write.csv(MySubmission, 'data/SummissionRandomForest.csv', row.names=F)
```
