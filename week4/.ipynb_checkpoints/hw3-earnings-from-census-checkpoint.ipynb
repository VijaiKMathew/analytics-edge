```r
library(caTools)
data = read.csv('data/census.csv.xz')
```

```r
head(data)
```

```r
set.seed(2000)
spl = sample.split(data$over50k,
                   SplitRatio = 0.6)
train = subset(data, spl==T)
test = subset(data, spl==F)
```

```r
lin1 = glm(over50k ~ ., data=train,
           family='binomial')
summary(lin1)
```

```r
pred = predict(lin1, newdata=test,
               type='response')
t = table(test$over50k, pred >= 0.5)
t
sum(diag(t)) / sum(t)
```

```r
t1 = table(test$over50k)
t1
max(t1)/sum(t1)
```

```r
library(ROCR)
ROCRpred = prediction(pred, test$over50k)
as.numeric(performance(ROCRpred, 'auc')@y.values)
```

```r
library(rpart)
library(rpart.plot)
```

```r
tree = rpart(over50k ~ .,
             data=train, method='class')
#pred = predict(tree, newdata = test,
#               type='class')
options(repr.plot.width=7, repr.plot.height=7)
prp(tree)
```

```r
pred = predict(tree, newdata = test,
               type='class')
t = table(test$over50k, pred)
t
sum(diag(t))/sum(t)
```

```r
library(ROCR)
prob = predict(tree, newdata = test)[,2]
ROCRpred = prediction(prob, test$over50k)
ROCRperf = performance(ROCRpred, 'tpr', 'fpr')
as.numeric(performance(ROCRpred, 'auc')@y.values)
# plot ROC AUC curve
options(repr.plot.width=4, repr.plot.height=4)
plot(ROCRperf, colorize=T)
```

```r
set.seed(1)
trainSmall = train[sample(nrow(train), 2000),]
```

```r
library(randomForest)
```

```r
set.seed(1)
forest = randomForest(over50k ~ .,
                      data=trainSmall)
pred = predict(forest, newdata = test)
t = table(test$over50k, pred)
t
sum(diag(t))/sum(t)
```

```r
vu = varUsed(forest, count=T)
vusorted = sort(vu, decreasing = F,
                index.return=T)
dotchart(vusorted$x,
         names(forest$forest$xlevels[vusorted$ix]))
```

```r
varImpPlot(forest)
```
