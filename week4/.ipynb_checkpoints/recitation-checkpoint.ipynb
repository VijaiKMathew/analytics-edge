```r
library(rpart)
library(rpart.plot)
```

```r
latlontree = rpart(MEDV ~ LAT + LON, data=boston)
prp(latlontree)
```

```r
par(mar=c(4,4,0,0.5)+0.1)
plot(boston$LON, boston$LAT)
points(boston$LON[boston$MEDV>=21.21],
       boston$LAT[boston$MEDV>=21.21],
       col='red', pch=19)
fittedvalues = predict(latlontree)
points(boston$LON[fittedvalues>=21.21],
       boston$LAT[fittedvalues>=21.21],
       col='blue', pch='$')
```

```r
library(caTools)
set.seed(123)
```

```r
split = sample.split(boston$MEDV, SplitRatio = 0.7)
train = subset(boston, split==T)
test = subset(boston, split==F)
```

```r
linreg = lm(MEDV ~ LAT + LON + CRIM + ZN + INDUS +
            CHAS + RM + AGE + DIS + RAD + TAX +
            PTRATIO,
            data=train)
```

```r
summary(linreg)
```

```r
linreg.pred = predict(linreg, newdata=test)
linreg.sse = sum((linreg.pred - test$MEDV)^2)
linreg.sse
```

```r
tree = rpart(MEDV ~ LAT + LON + CRIM + ZN + INDUS +
            CHAS + RM + AGE + DIS + RAD + TAX +
            PTRATIO,
            data=train)
```

```r
prp(tree)
```

```r
tree.pred = predict(tree, newdata=test)
tree.sse = sum((tree.pred - test$MEDV)^2)
tree.sse
```

```r
library(caret)
library(e1071)
```

```r
tr.control = trainControl(method='cv', number=10)
cp.grid = expand.grid(.cp = (0:10)*0.001)
tr = train(MEDV ~ LAT + LON + CRIM + ZN +
           INDUS + CHAS + NOX + RM + AGE +
           DIS + RAD + TAX + PTRATIO,
           data=train, method='rpart',
           trControl=tr.control,
           tuneGrid=cp.grid)
```

```r
tr
```

```r
best.tree = tr$finalModel
prp(best.tree)
```

```r
best.tree.pred = predict(best.tree, newdata=test)
best.tree.sse = sum((best.tree.pred - test$MEDV)^2)
best.tree.sse
```
